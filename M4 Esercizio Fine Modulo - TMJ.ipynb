```python
from matplotlib import pyplot as plt
```

```python
from plotly import express as px
stocks = px.data.stocks()
```

```python
import pandas as pd
```

```python
y = stocks.head().loc[: ,"GOOG"]
x = stocks.head().loc[: ,"date"]
plt.plot(x,y)
plt.title("ES 1 Google")
plt.ylabel("Value", color="Green")
plt.xlabel("Date", color="Blue")
```

```python

y = stocks.tail().loc[:,"GOOG"]
x = stocks.tail().loc[:,"date"]
plt.plot(x,y)
plt.title("ES 1 Google")
plt.ylabel("Value", color="Green")
plt.xlabel("Date", color="Blue")
```

```python
y = stocks.head(20).loc[:,"AAPL"]
x = stocks.head(20).loc[:,"date"]
plt.plot(x,y, color = "Red", linewidth = 2, linestyle = "--", marker = "o", markerfacecolor = "Black")

plt.title("Azioni Apple")
plt.ylabel("Valore", color="Green")
plt.xlabel("Data", color="Blue")
```

```python
stocks.plot()
plt.legend(loc=4)
```

```python
election = px.data.election()
election.loc[:, "Coderre"].sum()
```

```python
election.plot(kind="bar", y="total", x="district_id")
```

```python
election.head(4).plot(x="district", y=["Coderre","Bergeron","Joly"], kind="bar")
plt.savefig("Graf1.png", dpi=600)
```

```python
election.head(4).plot(x="district", y=["Coderre","Bergeron","Joly"], kind="bar", stacked="True")
plt.savefig("Graf2.png", dpi=600)
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.deck.value_counts()
#7 Ponti
```

```python
titanic.isnull().sum()
#Age = Media
#Deck/Town = Nuova Categoria "unknown"
```

```python
sns.jointplot(data=titanic, x="fare", y="age")
#Rapporto Eta'/Tasse, i 40enni hanno i soldi, il biglietto low cost va per la maggiore, fascia eta 18-30 +/- e' la maggioranza
```

```python
sns.countplot(data=titanic, x="class")
```

```python
sns.countplot(data=titanic, x="alive")
```

```python
sns.boxplot(data=titanic, x="fare")
#Lowcost per la maggiore
```

```python
sns.boxplot(data=titanic, x="age", y="fare")
```

```python
sns.swarmplot(data=titanic, x="age", y="fare")
```

```python
sns.kdeplot(data=titanic, x="age")
```

```python
sns.boxplot(data=titanic, x="survived", y="fare")
# $ = Vita Extra
```
